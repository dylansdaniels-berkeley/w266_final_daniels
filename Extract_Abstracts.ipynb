```python
# function to scrape title and abstract from a BioRxiv page

def scrape_biorxiv_abstract(url):
    """
    Scrapes the title and abstract from a BioRxiv page and returns them as a JSON record.

    Args:
        url: The URL of the BioRxiv page.

    Returns:
        A JSON record in the format:
        {
            "link": url,
            "title": "",
            "abstract": ""
        }
    """

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the title element
    title_element = soup.find('h1', class_='highwire-cite-title')
    title = title_element.text.strip() if title_element else ""

    # Find the abstract element
    abstract_element = soup.find('div', class_='abstract')
    abstract_paragraphs = abstract_element.find_all('p') if abstract_element else []
    abstract = "\n".join([p.text.strip() for p in abstract_paragraphs])

    # Create the JSON record
    record = {
        "link": url,
        "title": title,
        "abstract": abstract
    }

    return record

# example
url = "https://doi.org/10.1101/2023.10.25.564020"
record = scrape_biorxiv_abstract(url)

print(json.dumps(record, indent=4))
```

```python
# collocations of words

words_string = ' '.join(words)
words_string

def extract_context(text, words, window_size=5):
    data = []
    words = set(words)  # Convert to set for faster lookup
    words_found = set()

    # Split text into words using regular expression
    word_regex = re.compile(r'\b\w+\b')
    all_words = word_regex.findall(text)

    for i, word in enumerate(all_words):
        if word.lower() in words:
            words_found.add(word.lower())
            start_index = max(0, i - window_size)
            end_index = min(len(all_words), i + window_size + 1)
            context = ' '.join(all_words[start_index:end_index])
            data.append({'Word': word.lower(), 'Context': context})

    # Check if all words are found
    missing_words = words - words_found
    if missing_words:
        print(f"Warning: Some words not found in text: {', '.join(missing_words)}")

    return pd.DataFrame(data)

word_context = extract_context(
    words_string,
    list(top_words.index[:5]),
    window_size=2
)

word_context.sort_values('Word')
```

```python
def get_dois(search_url):
    dois = []

    # Send a GET request to the search URL
    response = requests.get(search_url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all spans with class 'doi_label'
        doi_labels = soup.find_all('span', class_='doi_label')

        # Extract DOIs from the spans and add them to the list
        for label in doi_labels:
            # Get the next sibling which contains the DOI
            doi = label.find_next_sibling(text=True).strip()
            dois.append(doi)

    return dois

# search term: "EEG Beta Event"
# this search has 45 (75-item max length) pages, for 3,318 total results
base_url = 'https://www.biorxiv.org/search/EEG%252BBeta%252BEvent%20numresults%3A75%20sort%3Arelevance-rank?'
# page_root = 'https://www.biorxiv.org/search/EEG%252BBeta%252BEvent%20numresults%3A75%20sort%3Arelevance-rank?page='

page_urls = [base_url + 'page=' + str(i) for i in range(2, 46)]
page_urls = [base_url] + page_urls
page_dois = []

for page in page_urls:
    page_dois.extend(get_dois(page))

print('\nNumber of articles scraped: ', len(page_dois))

# dois = get_dois(search_url)
# print("DOIs found on the page:", dois)
```

```python
# remove dois from page_dois that are in df already
new_page_dois = [doi for doi in page_dois if doi not in list(ft_articles['link'])]
```
